# Demographics & Lifestyle

### - How does age group affect annual medical costs and premiums.

```python
age = df.groupby('Age_group',observed=True)[['annual_medical_cost','annual_premium']].mean().reset_index()
age
```

```python
fig, ax = plt.subplots(1,2, figsize=(14,5))

sns.barplot(data = df, x ='Age_group', y = 'annual_medical_cost', ax =ax[0], palette = 'plasma',hue = 'Age_group', legend=False)
ax[0].set_title('Avg Annual Medical Cost vs Age Group')
ax[0].set_xlabel('Age Group')
ax[0].set_ylabel('Avg Medical Cost')


sns.barplot(data = df, x ='Age_group', y = 'annual_premium', ax =ax[1], palette = 'plasma',hue = 'Age_group', legend=False)
ax[1].set_title('Avg Annual Premium vs Age Group')
ax[1].set_xlabel('Age Group')
ax[1].set_ylabel('Avg Annual Premium')

plt.tight_layout()
plt.show()
```

#### CONCLUSION
- Medical costs and premiums increase steadily with age.
- The 0–19 age group has the lowest average medical cost and premium.
- The 80+ age group shows the highest values, with medical costs and premiums over.
- This trend reflects how healthcare needs and insurance pricing escalate with age, likely due to increased risk and utilization.


### - Do urban vs rural residents show differences in healthcare utilization?

```python

area = df.groupby('urban_rural',observed=True)[['visits_last_year','hospitalizations_last_3yrs']].mean().reset_index()
area
```

```python
fig, ax = plt.subplots(2,1, figsize=(20,15))

sns.barplot(data = df, x ='urban_rural', y = 'visits_last_year', ax =ax[0], palette = 'magma',hue = 'Age_group', legend=False,)
ax[0].set_title('last year visit by Area', fontsize = 20)
ax[0].set_xlabel('Area', fontsize = 16)
ax[0].set_ylabel('last year visit', fontsize = 16)
ax[0].tick_params(axis='x', labelsize=14)
ax[0].tick_params(axis='y', labelsize=14)


sns.barplot(data = df, x ='urban_rural', y = 'hospitalizations_last_3yrs', ax =ax[1], palette = 'magma',hue = 'Age_group', legend=False)
ax[1].set_title('hospitalization in last 3 years by Area', fontsize = 20)
ax[1].set_xlabel('Area', fontsize = 16)
ax[1].set_ylabel('hospitalization in last 3 years', fontsize = 16)
ax[1].tick_params(axis='x', labelsize=14)
ax[1].tick_params(axis='y', labelsize=14)


plt.tight_layout()
#plt.legend(loc = 'best')
plt.show()
```

#### CONCLUSION
- Urban and suburban residents tend to have more frequent doctor visits compared to rural residents, suggesting better access to outpatient care.
- Hospitalization rates over the last 3 years are higher in rural areas, which may reflect delayed treatment or limited preventive care.
- Across all areas, older age groups show increased utilization — both in visits and hospitalizations — highlighting age as a key driver of healthcare demand.
- These patterns suggest that location and age together influence healthcare access and intensity, with rural populations potentially facing more severe health episodes due to lower routine care.


### - Is income level linked to plan type or network tier?

```python
df.groupby(['plan_type','network_tier'])['income'].mean().reset_index()
```

```python
plt.figure(figsize=(20,5))
pivot = df.pivot_table(values='income', index='plan_type', columns='network_tier', aggfunc='mean')

sns.heatmap(pivot, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title('Income level to plan type and network tier')
plt.xlabel('Network Tier')
plt.ylabel('Plan Type')
plt.show()
```

#### CONCLUSION

- HMO (Health Maintenance Organization)
- PPO (Preferred Provider Organization)
- EPO (Exclusive Provider Organization)
- POS (Point of Service)
- Individuals with higher income levels tend to choose Gold and Platinum network tiers, especially within PPO and EPO plans.
- Bronze and Silver tiers are more common among lower-income groups, suggesting affordability plays a role in plan selection.
- Among plan types, POS and PPO show slightly higher average incomes, indicating these may be preferred by more financially secure individuals.
- The heatmap reveals that both plan type and network tier are influenced by income, which can guide insurers in designing tiered offerings that match customer segments.

### - Which procedure types drive the highest costs?

```python
pivot_table = data.pivot(
    index='proc_surgery_count',
    columns='proc_imaging_count',
    values='annual_medical_cost')

plt.figure(figsize=(8,6))
sns.heatmap(pivot_table, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("Average Annual Medical Cost by Procedure Type")
plt.xlabel("Imaging Procedure Count")
plt.ylabel("Surgery Procedure Count")
plt.show()
```

#### CONCLUSION
- Surgical procedures contribute more significantly to cost increases than imaging procedures.
- Combined high counts of both procedure types lead to the steepest cost spikes, with the highest observed cost reaching ₹18,748.
- Minimal intervention (0 surgeries, 0 imaging) corresponds to the lowest cost of ₹2,877.
- Missing values in the grid suggest rare combinations or data limitations, which should be acknowledged in further analysis.

### - Are high-risk individuals more likely to have higher premiums and claims paid?

```python
risk_summary = df.groupby('is_high_risk')[['annual_premium', 'total_claims_paid']].mean().reset_index()
risk_summary
```

```python
# Create a copy with readable labels
risk_summary_labeled = risk_summary.copy()
risk_summary_labeled['risk_label'] = risk_summary_labeled['is_high_risk'].map({0: 'No', 1: 'Yes'})

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Annual Premium by Risk Status
sns.barplot(data=risk_summary_labeled, x='risk_label', y='annual_premium', hue='risk_label', palette='RdBu', ax=axes[0])
axes[0].set_title('Mean Annual Premium by High-Risk Status')
axes[0].set_xlabel('High-Risk')
axes[0].set_ylabel('Mean Annual Premium')

# Total Claims Paid by Risk Status
sns.barplot(data=risk_summary_labeled, x='risk_label', y='total_claims_paid', hue='risk_label', palette='RdBu', ax=axes[1])
axes[1].set_title('Mean Total Claims Paid by High-Risk Status')
axes[1].set_xlabel('High-Risk')
axes[1].set_ylabel('Mean Total Claims Paid')

plt.tight_layout()
plt.show()
```

#### CONCLUSION
- Individuals marked as high-risk ("Yes") pay significantly higher annual premiums than those labeled "No", reflecting insurers’ pricing strategy to offset expected medical expenses.
- They also receive more in total claims paid, suggesting greater healthcare utilization or costlier treatments.
- This confirms that high-risk classification is a strong predictor of financial impact, both for the insurer and the insured.
